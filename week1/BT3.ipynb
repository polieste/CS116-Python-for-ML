# Bước 6

```python
from geopandas.tools import sjoin
```

```python
data = data.to_crs(epsg=4326)
```

```python
click_data = gpd.sjoin(data, click)
click_data
```

```python
import numpy as np
```

```python
temp = click_data.groupby(['Dist_Name']).count()
data['count'] =  np.array(temp['dbh'])
data
```

```python
top10 = data.sort_values('count', ascending = False)[0:10]
top10.reset_index(inplace = True)
top10_index = top10['index']
top10.drop(['index'], axis = 1, inplace = True)
top10
```

# Bước 7

```python
top10click = click_data.drop([x for x in range(24) if x not in np.array(top10_index)])
top10click
```

```python
x = top10click['point'].x
y = top10click['point'].y
```

```python
import numpy as np
X = np.c_[x, y]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=20, random_state=0).fit(X)
print('Centers found by scikit-learn:')
print(kmeans.cluster_centers_)
pred_label = kmeans.predict(X)
```

```python
pred_label
```

# Bước 8

Lưu cụm điểm nhiều nhất của mỗi quận huyện

```python
top10click['label'] = pred_label
top10click
```

```python
from shapely.geometry import Point
```

```python
df = {}
df['geometry'] = [Point(x[0], x[1]) for x in kmeans.cluster_centers_]
df['count'] = top10click.groupby(['label']).count()['dbh']
df = gpd.GeoDataFrame(df, crs = "EPSG:4326")
df['point'] = df['geometry']
df
```

```python
df = gpd.sjoin(data.drop(['count'], axis = 1), df).drop(['index_right'], axis = 1)
df.sort_values('count', ascending = False, inplace = True)
df
```

```python
df.reset_index(inplace = True)
```

```python
mySet = set()
for i in range(len(df)):
  if df['Dist_Name'][i] in mySet:
    df.drop(i, inplace = True)
  else:
    mySet.add(df['Dist_Name'][i])
```
